## Data cleaning

### Remove Duplicate Entries

```python
# Duplicate samples

a = andr_df.duplicated(subset=['App'])
b = ios_df.duplicated(subset=['track_name'])
print(andr_df[a][['App']])
print(ios_df[b][['track_name']])
```

```python
andr_df = andr_df.drop_duplicates(subset=['App']).reset_index(drop=True)
ios_df = ios_df.drop_duplicates(subset=['track_name']).reset_index(drop=True)
```

### Remove Non-English apps

For the purpose of this project, all the applications that are not meant for English-speaking users will also be removed. In order to reduce data loss, only application with its name containing more than 3 non-ASCII will be removed.

---

Vì mục đích của dự án này, tất cả các ứng dụng không dành cho người dùng nói tiếng Anh cũng sẽ bị xóa. Để giảm mất dữ liệu, chỉ ứng dụng có tên chứa nhiều hơn 3 ký tự không phải ASCII mới bị xóa.

```python
def has_non_ascii_chars(s):
    non_ascii_count = sum(1 for char in s if ord(char) > 127)
    return non_ascii_count > 3
```

```python
has_non_ascii_chars('My God :)')
```

```python
filtered_andr_df = andr_df[~andr_df['App'].apply(has_non_ascii_chars)].reset_index(drop=True)
filtered_ios_df = ios_df[~ios_df['track_name'].apply(has_non_ascii_chars)].reset_index(drop=True)
```

```python
filtered_andr_df.shape, filtered_ios_df.shape
```

## Analyzing data

As mentioned from the beginning, the aim is to determine type of applications that are likely to attract mỏe users since revenue heavily depends on the amount of users using these applications. To minimize overhead and risks, the validation strategy for an app idea consists of 3 steps:

1.   Build a minimal Android version of the app, and add it to Google Play store.
2.   If the app has good responses from users, the app will be kept developing.
3.   If the app is profitable after 6 months, an iOS version of app will be built and add to App Store.

It is important that the app profile should be successful on both Google Play and App Store markets.

---

Như đã đề cập từ đầu, mục đích là xác định loại ứng dụng có khả năng thu hút người dùng nhiều hơn vì doanh thu phụ thuộc rất nhiều vào lượng người dùng sử dụng các ứng dụng này. Để giảm thiểu chi phí và rủi ro, chiến lược xác thực ý tưởng ứng dụng bao gồm 3 bước:


1.   Xây dựng phiên bản Android tối thiểu của ứng dụng và thêm nó vào cửa hàng Google Play.
2.   Nếu ứng dụng có phản hồi tốt từ người dùng, ứng dụng sẽ tiếp tục phát triển.
3.   Nếu ứng dụng có lãi sau 6 tháng, phiên bản ứng dụng iOS sẽ được xây dựng và thêm vào App Store.

Điều quan trọng là hồ sơ ứng dụng phải thành công trên cả thị trường Google Play và App Store.

```python
def cvt_installs(series):
    return float(series.replace(',','').strip('+'))
```

```python
rating_count_total_andrapp = free_appandr_df.groupby('Category')['Installs'].sum()
```

```python
rating_count_total_andrapp
```

```python
category_andr_counts = free_appandr_df['Category'].value_counts()
```

```python
avg_n_installs = rating_count_total_andrapp / category_andr_counts
avg_n_installs.sort_values(ascending=False)
```
